```python
import pandas as pd
import numpy as np
import sklearn.model_selection as sk
from sklearn.linear_model import LinearRegression
from pandas import get_dummies
import seaborn as sns
```

```python
df=pd.read_csv('final_data.csv')
df=df.drop(columns=["Unnamed: 0"])
```

```python
y=df['Persistency_Flag']
x=df.drop(columns="Persistency_Flag")
x.head()
```

```python
x = get_dummies(data=x)
```

```python
x['Dexa_Freq_During_Rx']=x['Dexa_Freq_During_Rx'].astype(int)
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=sk.train_test_split(x,y,train_size=0.7)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
```

```python
for i in range (len(pred)):
  pred[i]=round(pred[i])
pred=pred.astype(int)
y_test=np.array(y_test)
```

```python
sum=0
for i in range(len(pred)):
   if pred[i]==y_test[i]:
      sum+=1
accuracy=sum/len(pred)*100
print(accuracy)
```
