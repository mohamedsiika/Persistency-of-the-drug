```python
import pandas as pd
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
import sklearn.model_selection as sk
from pandas import get_dummies
from sklearn.ensemble import GradientBoostingClassifier
```

```python
df=pd.read_csv('final_data.csv')
df=df.drop(columns=['Unnamed: 0','Risk_Untreated_Early_Menopause','Risk_Chronic_Liver_Disease','Risk_Estrogen_Deficiency','Risk_Immobilization'])
```

```python
y=df['Persistency_Flag']
x=df.drop(columns="Persistency_Flag")
x.head()
```

```python
x = get_dummies(data=x)
x['Dexa_Freq_During_Rx']=x['Dexa_Freq_During_Rx'].astype(int)
```

```python
x_train,x_test,y_train,y_test=sk.train_test_split(x,y)
y_test=np.array(y_test)
```

```python
model1 = AdaBoostClassifier(n_estimators=250,learning_rate=1)
model1.fit(x_train,y_train)
pred1=model1.predict(x_test)
```

```python
model2 = GradientBoostingClassifier(n_estimators=250,learning_rate=0.1)
model2.fit(x_train,y_train)
pred2=model2.predict(x_test)
```

```python
def accuracy(pred):
  sum=0
  for i in range(len(pred)):
    if pred[i]==y_test[i]:
        sum+=1
  accuracy=sum/len(pred)*100
  print(accuracy)
```

```python
accuracy(pred1)
accuracy(pred2)
```
